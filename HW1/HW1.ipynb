#### 2. Is This Poisson?

```python
HomeScoreMean = HomeScore.mean()
AwayScoreMean = AwayScore.mean()
```

```python
HomePoisson = np.random.poisson(lam = HomeScoreMean, size = len(HomeScore))
AwayPoisson = np.random.poisson(lam = AwayScoreMean, size = len(HomeScore))
PlotData = pd.DataFrame(data = {'HomeScore': HomeScore, 'HomePoisson': HomePoisson,'AwayScore': AwayScore, 'AwayPoisson': AwayPoisson})
```

```python
from plotnine import *
```

```python
ggplot(PlotData) +  geom_bar(aes(x = "HomeScore"),stat = 'count',colour = "green",fill = "green")+ geom_line(aes(x = "HomePoisson"),stat = 'count',colour = "r") + geom_point(aes(x = "HomePoisson"),stat = 'count',colour = "r") +\
labs(title = "Home Score vs Poisson Generated Score") + xlab("Score") + ylab("Number of Games")
```

```python
ggplot(PlotData) +  geom_bar(aes(x = "AwayScore"),stat = 'count',colour = "darkblue",fill = "darkblue")+ geom_line(aes(x = "AwayPoisson"),stat = 'count',colour = "r") + geom_point(aes(x = "AwayPoisson"),stat = 'count',colour = "r") +\
labs(title = "Away Score vs Poisson Generated Score") + xlab("Score") + ylab("Number of Games")
```

- Looking at the scores, we can say both home and away scores fit the poisson distribution with corresponding mean values.

## Task 2

```python
Task2Data = Bets.copy()
```

```python
# keep the latest bets of each bookmaker - match pair
Task2Data = Task2Data.merge(how = "inner", right = Task2Data.groupby(["match_id","odd_bookmakers"]).agg({"odd_epoch":"max"}), on = ["match_id","odd_bookmakers","odd_epoch"])
```

```python
# convert epoch to datetime
Task2Data.odd_epoch = pd.to_datetime(Task2Data.odd_epoch, unit='s')
```

- I chose the top 4 bookmakers with highest count in Bets data.

```python
ChosenBookmakers = Task2Data.groupby('odd_bookmakers').count()[["value"]].sort_values(by = "value",ascending = False)[0:4].index
Task2Data = Task2Data[Task2Data.odd_bookmakers.isin(ChosenBookmakers)]
```

```python
# filter variable with odd_* only
Task2Data = Task2Data[Task2Data.variable.isin(['odd_1',"odd_x","odd_2"])]
```

#### 1. Calculate the P(home win), P(tie) and P(away win) by P(x) = 1/odd

```python
Task2Data = Task2Data.drop(columns = ["odd_epoch"])

# pivot of data
Task2Data["Key"] = Task2Data["odd_bookmakers"] + "-" + Task2Data["match_id"].map(str)
Task2Data = Task2Data[["Key","odd_bookmakers","match_id"]].drop_duplicates().merge(right = Task2Data.pivot(index = "Key",columns = "variable", values = "value"), how = "inner", on = "Key").drop(columns = "Key") 
```

```python
Task2Data["prob_1"] = 1/Task2Data["odd_1"]
Task2Data["prob_x"] = 1/Task2Data["odd_x"]
Task2Data["prob_2"] = 1/Task2Data["odd_2"]
```

```python
# iff any of odds is 0, then corresponding probability will be inf. So I change those probabilities to 0.0001
Task2Data.loc[Task2Data.odd_1 == 0,"prob_1"] = 0.0001
Task2Data.loc[Task2Data.odd_x == 0,"prob_x"] = 0.0001
Task2Data.loc[Task2Data.odd_2 == 0,"prob_2"] = 0.0001
```

```python
Task2Data.head()
```

#### 2. Normalization of Probabilities

```python
Task2Data["prob_sum"] = Task2Data["prob_1"] + Task2Data["prob_x"] + Task2Data["prob_2"]
Task2Data["norm_prob_1"] = Task2Data["prob_1"] / Task2Data["prob_sum"]
Task2Data["norm_prob_x"] = Task2Data["prob_x"] / Task2Data["prob_sum"]
Task2Data["norm_prob_2"] = Task2Data["prob_2"] / Task2Data["prob_sum"]
Task2Data["phome_paway"] = Task2Data["norm_prob_1"] - Task2Data["norm_prob_2"]
```

```python
Task2Data.head()
```

#### 3. Plotting P(home win) - P(away win)

```python
# creating bins from -1 to 1 with 0.05 step
bins = np.arange(-1.0, 1.05, 0.05)

# in order to plot correctly (x axis will be numerical), i calculated the mid point of each bin
CutMids = pd.DataFrame(data = {'bin_ph_pa' : pd.cut(x = bins, bins = bins).categories.astype(str), 'midpoints' : np.arange(-0.975,1.025,0.05)})

# bins of p(home win) - p(away win)
Task2Data["bin_ph_pa"] = pd.cut(x = Task2Data.phome_paway, bins = bins).astype(str)

# join numerical mid points to bins
Task2Data = Task2Data.merge(right = CutMids, on = "bin_ph_pa", how = "left")
```

```python
# bring match results (1 if draw, else 0 -- so that we can sum them up to count # of draws)
# only finished matches
Draws = Matches.loc[Matches.match_status == 'Finished',["match_id","match_hometeam_score","match_awayteam_score"]]
Draws["IsDraw"] = np.where(Draws.match_hometeam_score == Draws.match_awayteam_score,1,0) 
Draws = Draws[['match_id','IsDraw']]
Task2Data = Task2Data.merge(right = Draws, how = "inner", on = "match_id")
```

```python
# calculate number of matches and number of draws for each bin
Task2Data["bin_match_count"] = Task2Data.groupby(['bin_ph_pa'])["bin_ph_pa"].transform('count')
Task2Data["bin_draw_count"] = Task2Data.groupby(['bin_ph_pa'])["IsDraw"].transform('sum')

# calculate probability of draw by # of draw / # of matches
Task2Data["bin_prob_x"] = Task2Data["bin_draw_count"] / Task2Data["bin_match_count"]
```

```python
Task2Data.head()
```

```python
p_task2 = ggplot(data = Task2Data) + geom_point(aes(x = "phome_paway",y = "norm_prob_x"),colour = "blue",size = 0.2) + geom_point(aes(x = "midpoints",y = "bin_prob_x"),colour = "red",size = 1) + \
xlab("P(Home Win) − P(Away Win)") + ylab("P(Draw)") + scale_x_continuous(breaks = [-1,-0.75,-0.5,-0.25,0,0.25,0.5,0.75,1]) + ylim(0,0.5)
display(p_task2)
```

- Bins where P(Home Win) − P(Away Win) is around (-0.3,-0.2) might have potential to earn money in the long run. Since occurence rate of draws in those bins, higher than the bookmakers odds.
